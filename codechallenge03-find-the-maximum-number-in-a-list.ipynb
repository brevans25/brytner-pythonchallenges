# Find the Maximum Number in a List

### **Problem**

Write a function that takes a list of numbers and returns the **maximum value** in the list — the largest number.

### **Example**
`find_max([4, 9, 1, 17, 2])` *will output: `17`* <br>
`find_max([-5, -9, -2, -12])` *will output: `-2`* <br>
`find_max([42])` *will output: `42`*

### *Task*

1. Define a function called `find_max` that accepts a list of integers or floats.

2. Return the largest number in the list.

3. **Do not** use the built-in `max()` function — try to solve it manually using:
    - A `for` loop to check each number
    - A variable to track the current maximum

### **Concepts Tested**

- Looping through a list
- Comparison and conditional logic
- List handling
- Defensive programming *(optional)*


### **Stretch Ideas**

- Add error handling for empty lists
- Allow the function to work with mixed positive and negative values
- Return both the max value and its index in the list
- Rewrite it using a while loop instead of a for loop

***Hint***

*You’ll need a variable to hold the current highest number as you loop through the list and update it whenever you find a bigger one.*



```python
def find_max(listOfNumbers):

    if listOfNumbers == []:
        print("Not a valid list!")
        return None

    else:
        currentMax = listOfNumbers[0]

        for eachNumber in listOfNumbers:
            if currentMax < eachNumber:
                currentMax = eachNumber 
        print(f'The max value in this list is {currentMax}!')

    return currentMax

find_max([])
find_max([4, 9, 1, 17, 2])
find_max([10, -3, 7, -20, 5])
find_max([100, 85, 302, 14, 67, 99, 250])
find_max([3, 3, 3, 3, 3])
find_max([2.5, 3.7, 0.1, -9.4, 3.699])
find_max([0, -1, -2, -3])
```
