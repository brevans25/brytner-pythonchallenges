# Check for a Palindrome

### **Problem**

Write a function that checks whether a given string is a palindrome â€” a word or phrase that reads the same backward as forward (ignoring capitalization, spaces, and punctuation).

### **Example**
`is_palindrome("racecar")` *will output: `True` * <br>
`is_palindrome("hello")` *will output: `False`* <br>
`is_palindrome("A man a plan a canal Panama")` *will output: `True`*

### *Task*

1. Define a function called `is_palindrome` that accepts a single string.

2. Return `True` if the string is a palindrome, and `False` otherwise.

3. Normalize the string by:
    - Converting it to lowercase
    - Removing spaces (and optionally punctuation)

4. Reverse the normalized string and compare it to the original normalized version.

***Tip**: Try using string slicing as a first approach.*

**Bonus:** After get it working, write a second version using a for loop to build the reversed string manually.

### **Concepts Tested**

- String normalization
- Comparison logic
- Function design and return values
- Function definition and return values
- Optional: working with Python's re (regular expression) module for bonus cleanup

### Stretch Ideas

- Use regular expressions to remove all non-alphanumeric characters
- Adapt the function to check if a number is a palindrome
- Return a message instead of a Boolean (e.g., "Yes, it's a palindrome!")
- Make the function case-insensitive without altering the original input

```python
def is_palindrome():
    userInput = input("Enter a word or phrase to check if it's a palindrome: ").lower().replace(" ", "") # .replace() to remove any whitespace

    reversedInput = ""

    for eachCharacter in userInput:
        reversedInput = eachCharacter + reversedInput
    
    if reversedInput == userInput:
        print("Your selection is a palindrome!")
    else:
        print("Not a palindrome")

is_palindrome()
```

### Bonus Challenge

Use regular expressions to remove all non-alphanumeric characters

```python
import re

def is_palindrome():
    
    userInput = input("Enter a word or phrase to check if it's a palindrome: ")

    cleanInput = re.sub("[^0-9a-zA-Z]+",'', userInput.lower()) 

    reversedInput = ""

    for eachCharacter in cleanInput:
        reversedInput = eachCharacter + reversedInput
    
    if reversedInput == cleanInput:
        print("Your selection is a palindrome!")
    else:
        print("Not a palindrome")

is_palindrome()
```
