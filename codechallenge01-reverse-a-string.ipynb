# Reverse a String

### **Problem**

Write a function that takes a single string as input and returns that string in reverse order.

### **Example**
`reverse_string("hello")` *will output: `olleh`* <br>
`reverse_string("Python")` *will output: `nohtyP`*

### *Task*

1. Define a function called reverse_string that accepts one argument (a string).

2. Return the reversed version of the string.<br>

***Tip**: Try using string slicing as a first approach.*

**Bonus:** After get it working, write a second version using a for loop to build the reversed string manually.

### **Concepts Tested**

- String manipulation
- Indexing and slicing
- Looping through strings (if attempting the bonus)
- Function definition and return values
- Stretch Ideas
- Add a feature that ignores whitespace or punctuation
- Make the function work for lists as well as strings
- Compare the performance of slicing vs loop methods for long strings

```python
def reverse_string():
    userInput = input("Enter a word you want to reverse: ").strip()
    while userInput.isnumeric(): ## If user enters integer, it prompts the user to add a word => str
        userInput = input("Please, enter a word, not a number: ").strip()
    reverseInput = userInput[::-1]
    return print(f'The reverse order of the word you entered is {reverseInput}')

reverse_string()
```

### Bonus Challenge

```python
def another_reverse_string():
    anotherUserInput = input("Enter a word you want to reverse: ").strip()
    while anotherUserInput.isnumeric(): ## If user enters integer, it prompts the user to add a word => str
        anotherUserInput = input("Please, enter a word, not a number: ").strip()

    returnReverse = "" # initializes an empty variable to store reversed string

    for eachLetter in anotherUserInput:
        returnReverse = eachLetter + returnReverse  # This is because as you update the variable returnReverse, you are adding the next character (eachLetter) in front of returnReverse:
                                                    # 1st iteration: gets 1st letter,
                                                    # 2nd iteration: gets 2nd letter and pushes 1st letter to the end... so on and so forth

    return print(f'The reverse order of the word you entered is {returnReverse}')

another_reverse_string()
```
